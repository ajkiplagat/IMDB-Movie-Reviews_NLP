```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
%matplotlib inline 
```

```python
import os
```

## a) Training set 

```python
# Create Data Frame with positive reviews
path = "IMDB/train/pos/"
temp = []
for file in os.listdir(path):
    with open(os.path.join(path + file), "r") as f:
        temp.append(f.readlines()[0])
df_pos = pd.DataFrame({"reviews": temp,"labels": 1})
df_pos.head()
```

```python
# Create Data Frame with negative reviews
path = "IMDB/train/neg/"
temp = []
for file in os.listdir(path):
    with open(os.path.join(path + file), "r") as f:
        temp.append(f.readlines()[0])
df_neg = pd.DataFrame({"reviews": temp,"labels": 0})
df_neg.head()
```

```python
# Vertically merge positive and negative reviews into one data frame
df = pd.concat([df_pos, df_neg], ignore_index = True)
#df
```

```python
df.head()
```

```python
df.info()
```

```python
# Shuffle rows to mix up negative and positive reviews
df = df.reindex(np.random.permutation(df.index))
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
# Export training set to csv 
df.to_csv('train_set.csv', index = False)
```
